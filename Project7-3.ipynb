# Question 1
## Part A
Cov(X) has two linearly independent vectors, so X has two principal components. 

## Part B

```
import numpy as np

pY0 = 0.5
pY1 = 0.5
E0 = np.array([[1,1],[1,2]])
E1 = np.array([[1,0],[0,1]])
C = (pY0 * E0) + (pY1 * E1)
print("Cov(X) = ")
print(C)
```

## Part C

```
print("\nEigen values = ")
eigenVals, eigenVecs = np.linalg.eig(C)
print(eigenVals)
print("\nEigen vector #1", eigenVecs[:,0])
print("\nEigen vector #2",eigenVecs[:,1])
```

## Part D

10 samples

```
import matplotlib.pyplot as plt
x1, x2 = np.random.multivariate_normal([0,0], C, 10).T
plt.plot(x1, x2, 'x')
plt.axis('equal')
C10 = np.cov(x1, x2)
print("Cov(X) = ")
print(C10)
print("\nEigen values = ")
eigenVals, eigenVecs = np.linalg.eig(C10)
print(eigenVals)
# Scale the unit eigen vectors by their corresponding eigen values
vec1 = eigenVals[0] * eigenVecs[:,0]
vec2 = eigenVals[1] * eigenVecs[:,1]
plt.plot([0, vec1[0]], [0, vec1[1]], linewidth = 2)
plt.plot([0, vec2[0]], [0, vec2[1]], linewidth = 2)
plt.xlabel('x1')
plt.ylabel('x2')
plt.show()
```

## Part E

```
diff = [];
for i in range(10000):
  x1, x2 = np.random.multivariate_normal([0,0], C, i).T
  tempC = np.cov(x1, x2)
  diff.append(np.linalg.norm(C - tempC))
plt.plot(diff)
plt.xlabel('Number of Samples')
plt.ylabel('Cov(X) - sample Cov(X)')
plt.title('Difference between true Cov(X) and sample Cov(X)')
```

## Part F
As we use infinitely many sample points, the sample covariance matrix will closer resemble the true covariance matrix as demonstrated from the graph above. The plot depicts the decreasing difference between the true covariance matrix and the sample covariance matrix as we increase the number of samples (stopped at 10,000 sample for sake of run time). This trend is an example of the law of large numbers.

## Part A

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from scipy.spatial import distance
%matplotlib inline
# Download data 
gmmLink = 'https://raw.githubusercontent.com/ds4ph-bme/class-project-7-skhan102jh/main/gmm.csv?token=AP2OF2LXESVRTYVMSAKAPJC7TIO4E'
swissLink= 'https://raw.githubusercontent.com/ds4ph-bme/class-project-7-skhan102jh/main/swissroll.csv?token=AP2OF2ILEG36Z242Y5WBGIS7TIO6M'

# Read data into data frames
gmm = pd.read_csv(gmmLink)
swiss = pd.read_csv(swissLink)

# Drop unnecessary columns
gmm = gmm.drop(gmm.columns[0], axis = 1)
swiss = swiss.drop(swiss.columns[0], axis = 1)

# We want to determine the best k to use for each of the data sets
# gmm elbow test
distort = [] 
gmm_array = gmm.to_numpy()
kVals = range(1, 20)
for i in kVals: 
  tempKmodel = KMeans(n_clusters = i).fit(gmm_array)
  tempKmodel.fit(gmm_array)
  distort.append(sum(np.min(distance.cdist(gmm_array, tempKmodel.cluster_centers_, 'euclidean'), axis = 1)) / gmm_array.shape[0])
plt.figure()
plt.plot(kVals, distort, 'bo-')
plt.title('Elbow test for gmm data set')
plt.xlabel('k value')
plt.ylabel('distortion')
# swiss roll elbow test
distort = [] 
swiss_array = swiss.to_numpy()
kVals = range(1, 20)
for i in kVals: 
  tempKmodel = KMeans(n_clusters = i).fit(swiss_array)
  tempKmodel.fit(swiss_array)
  distort.append(sum(np.min(distance.cdist(swiss_array, tempKmodel.cluster_centers_, 'euclidean'), axis = 1)) / swiss_array.shape[0])
plt.figure()
plt.plot(kVals, distort, 'bo-')
plt.title('Elbow test for swissroll data set')
plt.xlabel('k value')
plt.ylabel('distortion')

# We want 4 clusters for gmm
KmodelGMM = KMeans(n_clusters = 4).fit(gmm)
plt.figure()
plt.title('gmm.csv')
plt.scatter(gmm['X_1'].values, gmm['X_2'].values, c = KmodelGMM.labels_.astype(float))

# We want 2 clusters for swissroll
KmodelSW = KMeans(n_clusters = 5).fit(swiss)
plt.figure()
plt.title('swissroll.csv')
plt.scatter(swiss['X_1'].values, swiss['X_2'].values, c = KmodelSW.labels_.astype(float))
```

## Part B
For each data set I performed an elbow test to see if there is a k value that is most suitable. For the gmm data set, there is a clear elbow at k = 4, so I chose to choose the number of clusters to be 4. For the swissroll data set, the elbow in the plot is less obvious; however, there still seems to be the most drastic change in the slope steepness at k = 5. Thus, I chose k = 5 for the second plot.

## Part C
k-means clustering worked very well on the first data set in separating the data into 4 categories. We can verify that this method fo categorization is effective as we can visually see that there are four spatially separate groups of data in the gmm data set. k-means clustering was not suitable to be used for the second data set since the data does not appear to have spatially separate groups of data. In order to better categorize the swissroll data, we could take advantage of its concentric/spiraling circles and separate the data into categories based on their radius from the center of the spiral at (0, 0).
